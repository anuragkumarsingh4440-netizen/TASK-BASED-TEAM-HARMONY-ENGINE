 # Purpose:
Visualize top team and solo recommendations from 05_team_recommender.ipynb using simple, interactive charts.




### "I used Matplotlib and Seaborn to visualize the results of team recommendations and solo performers for the HR dashboard."

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load team recommendations
df_teams = pd.read_csv("top_team_recommendations.csv")
df_solo = pd.read_csv("top_solo_recommendations.csv")

# Show top 5 teams
top_teams = df_teams.head(5)

#  Bar chart: Skill vs Synergy vs Total Score
plt.figure(figsize=(12, 6))
team_labels = top_teams["Team Members"]

x = range(len(team_labels))
plt.bar(x, top_teams["Skill Score"], width=0.25, label="Skill Score")
plt.bar([i + 0.25 for i in x], top_teams["Synergy Score"], width=0.25, label="Synergy Score")
plt.bar([i + 0.5 for i in x], top_teams["Total Score"], width=0.25, label="Total Score")

plt.xticks([i + 0.25 for i in x], team_labels, rotation=45, ha='right')
plt.title("Top 5 Team Recommendations")
plt.ylabel("Score")
plt.legend()
plt.tight_layout()
plt.show()

#  Top Solo Performers
print(" Top Solo Performers:")
print(df_solo[["Employee Name", "Match Score"]].head(5))
```

### "I created this 2nd visualization using the Plotly library, based on the results of team recommendations and solo performers, for inclusion in the HR dashboard."

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px

# Load team recommendations
df_teams = pd.read_csv("top_team_recommendations.csv")
df_solo = pd.read_csv("top_solo_recommendations.csv")

# Show top 5 teams
top_teams = df_teams.head(5)

#  Interactive Bar Chart: Skill, Synergy, Total
fig = go.Figure()

fig.add_trace(go.Bar(
    x=top_teams["Team Members"],
    y=top_teams["Skill Score"],
    name="Skill Score",
    marker_color='lightskyblue'
))

fig.add_trace(go.Bar(
    x=top_teams["Team Members"],
    y=top_teams["Synergy Score"],
    name="Synergy Score",
    marker_color='lightgreen'
))

fig.add_trace(go.Bar(
    x=top_teams["Team Members"],
    y=top_teams["Total Score"],
    name="Total Score",
    marker_color='salmon'
))

fig.update_layout(
    title="Top 5 Team Recommendations",
    xaxis_title="Team Members",
    yaxis_title="Score",
    barmode='group',
    hovermode='x unified',
    template='plotly_white'
)

fig.show()

#  Interactive Table: Top Solo Performers
fig_solo = px.bar(
    df_solo.head(5),
    x="Employee Name",
    y="Match Score",
    title="Top Solo Performers",
    color="Match Score",
    text="Match Score",
    template="plotly_white"
)

fig_solo.update_traces(textposition="outside")
fig_solo.show()
```
